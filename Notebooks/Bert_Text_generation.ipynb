```python
#Usage of deberta model
#path - D:\Projects\LLM-AI-generated-Text-Detection\model_saves\deberta
#it contains 7 files - 
#1. config.json
#2. added_tokens.json
#3. special_tokens_map.json
#4. tokenizer_config.json
#5. tokenizer.json
#6. model.safetensors
#7. spm.model
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Load the model and tokenizer
model_path = path  # Using the path from the previous cell
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path)

# Function to predict if text is AI-generated
def predict_ai_text(text):
    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
    
    # Make prediction
    with torch.no_grad():
        outputs = model(**inputs)
    
    # Get probabilities
    probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
    
    # Assuming binary classification (AI-generated or not)
    ai_prob = probs[0][1].item()
    
    return ai_prob

# Example usage
sample_text = "This is a sample text to test the AI detection model."
ai_probability = predict_ai_text(sample_text)
print(f"Probability of being AI-generated: {ai_probability:.2%}")
```
